```python
import csv
```

```python
# Stitch the dismabiguated relations in with the relations

disambiguated = []

with open('data/csv/disambiguated-relation.csv', encoding='utf-8-sig') as file:
    reader = csv.DictReader(file)
    for row in reader:
        disambiguated.append(row)
    file.close()

relations = []

with open('data/csv/relation.csv', encoding='utf-8-sig') as file:
    reader = csv.DictReader(file)
    for row in reader:
        relations.append(row)
    file.close()
```

```python
for rel in disambiguated:
    target_rel = [r for r in relations if r['relation_id'] == rel['relation_id']][0]
    try:
        target_rel['target_maker_id'] = rel['target_maker_id']
    except:
        continue
    
```

```python
with open('data/csv/relation2.csv', 'w') as file:
    fieldnames = list(relations[0].keys())
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for row in relations:
        writer.writerow(row)
    file.close()
```
